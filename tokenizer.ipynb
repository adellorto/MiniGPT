```python
import tiktoken

# Initialize tokenizer
enc = tiktoken.get_encoding("cl100k_base")

# Text to test
text = "hello world"

# Print original text
print("Original text:", text)

# Encode
tokens = enc.encode(text)
print("Encoded tokens:", tokens)

# Decode
decoded_text = enc.decode(tokens)
print("Decoded text:", decoded_text)
```

```python
# Character-level tokenizer example

# Text to test
text = "hello world"

# Build vocabulary (unique characters)
chars = sorted(list(set(text)))
stoi = {ch: i for i, ch in enumerate(chars)}  # char -> int
itos = {i: ch for i, ch in enumerate(chars)}  # int -> char

# Print original text
print("Original text:", text)

# Encode (string → list of ints)
encoded = [stoi[c] for c in text]
print("Encoded:", encoded)

# Decode (list of ints → string)
decoded = ''.join([itos[i] for i in encoded])
print("Decoded:", decoded)
```

```python
# Train a minbpe tokenizer on Lorem Ipsum and test encode/decode

from minbpe.minbpe.regex import RegexTokenizer

# 1. Training text (Lorem Ipsum)
text = """
Lorem ipsum dolor sit amet, consectetur adipiscing elit.
Sed do eiusmod tempor incididunt ut labore et dolore magna aliqua.
Ut enim ad minim veniam, quis nostrud exercitation ullamco laboris.
"""

# 2. Initialize tokenizer
tokenizer = RegexTokenizer()

# 3. Train tokenizer (choose vocab size)
tokenizer.train(text, vocab_size=300)

# 4. Test text
test_text = "hello world"

print("Original text:", test_text)

# 5. Encode
encoded = tokenizer.encode(test_text)
print("Encoded:", encoded)

# 6. Decode
decoded = tokenizer.decode(encoded)
print("Decoded:", decoded)
```
